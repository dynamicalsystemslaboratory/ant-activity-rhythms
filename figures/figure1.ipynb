```python
# Import and unpack colony-level data.
colony = 'EGTi'
colonyfilelist, (colonynum, Nlist), (dt, tlist, adata) = importcolonydata( colony )

# Filter out bad colonies from list.
colonybad = []#[1, 4, 8, 10, 13, 18, 23, 29, 32]
colonynum = np.setdiff1d( np.argsort( Nlist ), colonybad, assume_unique=True )
```

```python
jlist = [34]
idlist = '-'.join( [str( num ) for num in jlist] )

# Plot some of the activity time-series.
fig, axs = plt.subplots( len( jlist ), sharex=True )
axs = axs if len( jlist ) > 1 else [axs]

for i, j in enumerate( jlist ):
    axs[i].plot( tlist, adata[j], color='k' )
    if len( jlist ) > 1:
        axs[i].text( 0.95*tlist[-1], 0.90, '$N = %i$' % Nlist[j],
            ha='right', va='top' )

    axs[i].set_xlim( [0,800] )
    axs[i].set_ylim( [0,1] )

    axs[i].set_ylabel( '$a(t)$' if len( jlist ) > 1 else 'proportion of\nants active' )

axs[-1].set_xlabel( '$t$' if len( jlist ) > 1 else 'time (min)' )

fig.set_figwidth( 4 )
fig.set_figheight( 2*len( jlist ) )
fig.tight_layout()

if 0:
    savefigure( fig, 'experiments/' + colony + '-' + idlist + '_activity_time-series' )
```
